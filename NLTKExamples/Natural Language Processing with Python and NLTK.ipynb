```python
Text = """Apparently, Ernest Hemingway was lunching at Luchow's with a number of writers and claimed \
that he could write a short story that was only six words long. Of course, the other writers balked. \
Hemingway told each of them to put ten dollars in the middle of the table; if he was wrong, he said, \
he’d match it. If he was right, he would keep the entire pot. He quickly wrote six words down on a \
napkin and passed it around; Papa won the bet. The words were "FOR SALE, BABY SHOES, NEVER WORN." \
A beginning, a middle and an end!"""
```

# Tokanaizing Words and Sentenses
* In word tokenize, puntuation characters [.,?!;'"] identifies as a word.

```python
Sentences = nltk.sent_tokenize(Text)   
print(Sentences)
```

```python
Words = nltk.word_tokenize(Text)
print(Words)
```

# Stop words

```python
StopWords = set(nltk.corpus.stopwords.words('english'))
print(StopWords)
```

```python
import string
punctuations = list(string.punctuation)
punctuations
```

```python
ExcludeWordList = list(StopWords) + punctuations + ['``', "''", 'A', '’', "'s"] + ['AM', 'PM']
ExcludeWordList
```

```python
WordsFiltered = []

for w in Words:
    if w not in ExcludeWordList:
        WordsFiltered.append(w)
        
print(WordsFiltered)
```

# Stemming 

```python
PS = nltk.stem.PorterStemmer()

WordsStemmed = []

for w in Words:
    sw = PS.stem(w)
    WordsStemmed.append(sw)
    print('{} -> {}'.format(w,sw))

print(WordsStemmed)
```

# Part of Speech Tagging

```python
POSTags = nltk.pos_tag(Words)
print(POSTags)
```

# Analyze Webpage
## Tokanize words, Get word counts and generate a word cloud

```python
import urllib
from bs4 import BeautifulSoup
import re
import string
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
url = "https://en.wikipedia.org/wiki/Albert_Einstein"
html = urllib.request.urlopen(url).read()
soup = BeautifulSoup(html, "lxml")
```

```python
# Remove scripts and Styles from html file
for script in soup(["script", "style"]):
    script.decompose()    # rip it out

# Extract text
Text = soup.get_text()

# Remove extra line breaks and spaces
Text=re.sub('[\n][\n]+' , '\n', Text)
Text=re.sub('[ ][ ]+' , '\n', Text)
```

```python
StopWords = set(nltk.corpus.stopwords.words('english'))
Punctuations = list(string.punctuation)
ExcludeWordList = list(StopWords) + punctuations + ['``', "''", '’', "'s"] 
```

```python
Words = nltk.word_tokenize(Text)

WordsFiltered = []
for w in Words:
    if w.lower() not in ExcludeWordList:
        WordsFiltered.append(w)        
```

```python
WordCounter = Counter(WordsFiltered)
WordFrequency = pd.DataFrame(data = WordCounter.most_common(50), columns = ['Word', 'Frequency'])
WordFrequency
```

```python
WordFrequencyDict = dict(zip(WordFrequency['Word'].values, WordFrequency['Frequency'].values))
wc = WordCloud(max_words=100, margin=10, background_color="white").generate_from_frequencies(WordFrequencyDict)
```

```python
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
